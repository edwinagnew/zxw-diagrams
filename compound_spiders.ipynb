```python
from pyfile import eval, eq
from discopy.quantum.zx import Spider, Id
from spiders import W, Z, Swap, H, CZ, FSwap, X, ZBox

def T(A, n):
    # return A^n
    if n == 1: return A
    return A @ T(A, n-1)
```

```python
S22 = W() >> Z(1, 2) @ Z(1, 2)
S22.draw()
```

```python
P22 = Z(1, 2) >> W() @ W()
P22.draw()
```

```python
P22_alt = W(4) >> T(Z(1, 2), 4) >> Id(1) @ Swap @ Swap @ Swap @ Id(1) >> T(W(2).dagger(), 4) >> Id(2) @ Swap >> Id(1) @ Swap @ Id(1)
P22_alt.draw()
eq(P22, P22_alt)
```

```python
x = W() @ W() >> Z(1, 2) @ Z(2, 2) @ Z(1, 2)
x.draw()
eval(x).astype(float).nonzero()
```

```python
y = Z(1, 2) >> W() @ W() >> Id(1) @ Z(2, 1) @ Id(1)
y.draw()
eval(y).astype(float)
```

```python
y2 = P22_alt >> Id(1) @ Z(2, 1) @ Id(1)
y2.normal_form().draw()
eq(y, y2)
```

```python
y3 = W(4) >> Z(1, 3) @ Z(1, 2) @ Id(1) @ Z(1, 2) >> Id(2) @ Swap @ Id(1) @ Swap @ Id(1) >> Id(4) @ Swap @ Id(2) >> Id(1) @ Swap @ X(2, 0) @ Swap @ Id(1) >> W(2, False) @ W(2, False) @ W(2, False)
y3.normal_form().draw()
eq(y, y3)
```

```python
eval(y3).astype(float)
```
